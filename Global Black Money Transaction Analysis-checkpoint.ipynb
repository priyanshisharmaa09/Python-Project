## 2. Distribution of transaction amount

```python
plt.figure(figsize=(10,6))
sns.histplot(df["Amount (USD)"],bins=200,kde=True,color="Blue")
plt.title("Distribution of Transaction Amounts",)
plt.ylabel("Frequency")
plt.xlabel("Transaction Amount in USD")
plt.show()
```

The bar chart displays the Distribution of Transaction Amounts with a frequency count on the y-axis and transaction amounts (in USD) on the x-axis. Key insights from the chart:

1. Uniform Distribution: The transaction amounts are relatively uniformly distributed across different value ranges, indicating that transactions occur consistently across a wide range of amounts, from smaller to larger sums.

2. No Strong Peaks: There are no significant peaks or concentration at any particular transaction amount range, which suggests that there is no dominant amount frequently transacted.

3. Slight Drop at Extremes: There is a noticeable decline in frequency for transactions near the very lowest and highest ends of the spectrum. This is typical, as very small or very large transactions are generally less common compared to mid-range amounts.

4. Smooth Curve Overlay: The smooth line overlaid on the histogram shows a slight bell shape but remains relatively flat, confirming the broad and uniform distribution with a gentle decline at the higher and lower extremes.

In summary, this distribution suggests that the transaction amounts are evenly spread, without any particular bias towards certain values, except for fewer transactions at the extreme ends.

# Advanced Exploratory Analysis

## 1.Top 10 countries by Transaction Number and Amount of Transaction

```python
top_countries_transaction=df['Country'].value_counts().head(10)
top_countries_amount=df.groupby('Country')['Amount (USD)'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
sns.barplot(x=top_countries_transaction.index,y=top_countries_transaction.values,palette='coolwarm',hue=top_countries_transaction,legend=False)
plt.title("Top 10 Countries by Number of Transactions")
plt.xlabel("Country")
plt.ylabel("Number Of Transactions")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
sns.barplot(x=top_countries_amount.index,y=top_countries_amount.values,palette='Blues_r',hue=top_countries_amount,legend=False)
plt.title("Top 10 Countries by Transaction Amount")
plt.xlabel("Country")
plt.ylabel("Amount Of Transactions (USD)")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show
```

The two charts compare the top 10 countries by number of transactions (left) and top 10 countries by total transaction amount in USD (right). Here are the insights:

1. China, South Africa, and the UK appear at the top in both charts, indicating that these countries not only have a high number of transactions but also large total transaction values. This suggests that significant financial activities are occurring in these countries, both in terms of volume and monetary value.
   
2. The number of transactions is relatively consistent across the top 10 countries (left chart), with most countries performing just under or around 1,000 transactions. However, in the right chart, the total transaction amounts vary more significantly, with China having the highest total, followed closely by South Africa and the UK.

3. USA and UAE rank lower in both charts, meaning they have fewer transactions and lower total amounts compared to the other countries in the top 10.

4. Some countries, like Switzerland and India, have a similar number of transactions but contribute less to the total transaction amount. This could mean that their individual transaction sizes are smaller compared to countries like China or South Africa.

In summary, China leads both in transaction count and total amount, followed by South Africa and the UK. While the number of transactions is fairly even among the top 10, the total transaction values reveal larger differences, highlighting that some countries handle larger individual transaction amounts.

## 2.Time series analysis of transaction volume and transaction amount over time

```python
df.set_index('Date of Transaction',inplace=True)

monthly_transaction_count=df.resample('M').size()
monthly_transaction_amount=df['Amount (USD)'].resample('M').sum()

# Visualization: Transaction count and transaction amount over time
plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
monthly_transaction_count.plot(color='green', linewidth=2)
plt.title('Monthly Number of Transactions')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')


plt.subplot(2,1,2)
monthly_transaction_amount.plot(color='purple', linewidth=2)
plt.title('Monthly Total Transaction Amount (USD)')
plt.xlabel('Date')
plt.ylabel('Total Amount (USD)')

plt.tight_layout()
plt.show()
```

#### Monthly Number of Transactions (Top Chart):

1. The number of transactions remained relatively stable between January and December 2013, fluctuating around 650-750 transactions per month.

2. While there are minor monthly fluctuations, they are not significant. However, in January 2014, there is a dramatic drop in the number of transactions, nearly reaching zero.

3. This sharp decline suggests a possible event that caused transactions to halt suddenly at the beginning of 2014, potentially due to changes in policy, regulations, or external factors.

#### Monthly Total Transaction Amount (Bottom Chart):

1. The total transaction value per month mirrors the pattern of the number of transactions, with slight fluctuations between 1.6 billion and 1.9 billion throughout 2013.

2. There is no major shift in the total transaction amount until the sharp drop in January 2014, aligning with the decrease in the number of transactions

Conclusion: Throughout 2013, both the number and total value of transactions remained relatively stable with minor fluctuations. However, in early 2014, there was a significant drop in both metrics, indicating a major shift in the system or external environment affecting these transactions.

## 3.Correlation between transaction amount, money laundering risk score and shell companies involved

```python
plt.figure(figsize=(8,6))

sns.heatmap(df[['Amount (USD)','Money Laundering Risk Score', 'Shell Companies Involved']] .corr(),annot=True,vmin=-1,vmax=1,cmap='coolwarm')
plt.title('Correlation Matrix between Amount, Risk Score, and Shell Companies')
plt.show()
```

The correlation matrix provides insights into the relationships between the transaction amount (USD), money laundering risk score, and the involvement of shell companies. Here are the key insights:

1. Transaction Amount vs. Money Laundering Risk Score:
   
--The correlation coefficient is 0.017, which is very close to zero. This suggests that there is no meaningful relationship between the transaction amount and the money laundering risk score.

2. Transaction Amount vs. Shell Companies Involved:

--The correlation coefficient is -0.018, which indicates a very weak negative relationship. This means that the transaction amount has almost no influence on whether shell companies are involved or not.

3. Money Laundering Risk Score vs. Shell Companies Involved:

--The correlation coefficient is -0.019, also very close to zero. This suggests that there is no significant correlation between the money laundering risk score and the involvement of shell companies.


Conclusion: There are no strong correlations between the three variables (transaction amount, money laundering risk score, and shell company involvement), indicating that changes in one of these variables are not strongly associated with changes in the others. This could suggest that other factors not captured in this matrix may influence these metrics.

## 4.Industry-based Analysis : Total transaction amount per industry and Average risk score per industry

```python
industry_amount=df.groupby('Industry')['Amount (USD)'].sum().sort_values(ascending=False)
industry_risk=df.groupby('Industry')['Money Laundering Risk Score'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
sns.barplot(x=industry_amount.index , y=industry_amount.values , palette='magma' , hue=industry_amount.index , legend=False)
plt.title('Transaction Amount per Industry')
plt.xlabel('Industry')
plt.ylabel('Amount in USD')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
sns.barplot(x=industry_risk.index , y=industry_risk.values , palette='Reds' , hue=industry_risk.index , legend=False)
plt.title('Average Money Laundering Risk Score per Industry')
plt.xlabel('Industry')
plt.ylabel('Average Risk Score')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```

The two bar charts above provide insights into the total transaction amounts and the average money laundering risk scores across different industries. Here are the key insights:

1. Total Transaction Amount by Industry (Left Chart):
   
--The Finance industry has the highest total transaction amount, slightly exceeding 3.5 billion USD, indicating its significant economic activity and large volume of financial transactions.

--Construction and Arms Trade also have high total transaction amounts, followed closely by Luxury Goods and Real Estate industries.

--Casinos and Oil & Gas industries, while still substantial, have slightly lower transaction volumes compared to others.

2. Average Money Laundering Risk Score by Industry (Right Chart):
   
--The Finance industry also has the highest average risk score, which could suggest that the financial industry has a higher perceived risk for money laundering activities.

--Construction and Arms Trade have similarly elevated risk scores, indicating that these industries are also considered high-risk for money laundering.

--Casinos and Real Estate industries show moderate risk scores.

--Interestingly, Luxury Goods and Oil & Gas industries show slightly lower risk scores compared to others, but they still maintain relatively high average risk levels.

Conclusion: The Finance, Construction, and Arms Trade industries not only have high transaction amounts but also exhibit higher money laundering risk scores, highlighting their vulnerability to potential illicit activities. While the Oil & Gas and Luxury Goods sectors have significant economic activity, their risk scores are slightly lower in comparison, which could indicate a different risk profile or regulatory environment.



## 5.High-Risk Transactions: Transactions with a risk score above 8 

```python
high_risk_transactions=df[df['Money Laundering Risk Score']>8]

high_risk_industries=high_risk_transactions['Industry'].value_counts().head(5)

plt.figure(figsize=(8,6))
sns.barplot(x=high_risk_industries.index, y=high_risk_industries.values, palette='coolwarm' , hue=high_risk_industries.index , legend=False)
plt.title('Top 5 Industries Involved in High-Risk Transactions (Risk Score > 8)')
plt.xlabel('Industry')
plt.ylabel('Number of High-Risk Transactions')
plt.xticks(rotation=45)
plt.show()
```

1. Finance is the most prominent industry, accounting for the highest number of high-risk transactions, with over 320 occurrences. This suggests that the financial sector is highly exposed to money laundering activities or other illicit transactions, possibly due to the large volume and complexity of financial transactions.
   
2. Construction comes in second, with a slightly lower number of high-risk transactions, followed closely by Real Estate. These industries are known for their vulnerability to money laundering, often due to large, one-time payments and complex transaction chains.

   
3. Casinos and the Arms Trade also feature prominently, indicating that these sectors are frequently involved in high-risk financial activities, likely due to the nature of cash-intensive businesses (in the case of casinos) and the sensitive nature of arms trading.

## 6. Analyzing Shell Company Involvement: Transactions involving more than 2 shell companies

```python
shell_company_transactions=df[df['Shell Companies Involved']>2]

shell_company_per_country=shell_company_transactions['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=shell_company_per_country.index, y=shell_company_per_country.values, palette='Set1',hue=shell_company_per_country.index , legend=False)
plt.title('Top 10 Countries with Transactions Involving > 2 Shell Companies')
plt.xlabel('Country')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.show()
```

The bar chart shows the Top 10 Countries with Transactions Involving More Than 2 Shell Companies. Here are the insights:

1. China leads the list with over 700 transactions involving more than two shell companies, indicating a high frequency of complex corporate structures that could be associated with efforts to obscure financial activity or evade scrutiny.
   
2. Brazil, South Africa, Russia, and the UK follow closely, each with a similar number of transactions (just above 650). This highlights these countries as hotspots for transactions with potentially opaque ownership structures involving shell companies.

   
3. Other countries such as the UAE, Switzerland, the USA, Singapore, and India also show substantial activity in terms of transactions involving multiple shell companies, suggesting a widespread use of these entities across various jurisdictions.

   
Conclusion: These countries may require enhanced financial oversight and transparency measures, as transactions involving multiple shell companies are often associated with risks like money laundering, tax evasion, and other illicit financial activities.
